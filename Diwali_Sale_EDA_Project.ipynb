#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
null_values = df.isna()== True
null_values
#we have check the null values in DataFrame
````

####Drop Unnecessary Columns

````python
df.drop(columns=['Status', 'unnamed1'], inplace=True, errors='ignore')
````

####Handle Missing Values

````python
a = df.isnull().sum()
missing_values=a.sort_values(ascending = False)[:5]
missing_values
````

````python
plt.figure(figsize=(10,6))
missing_values.plot(kind = 'bar', fontsize = 20,legend='True')
plt.title('Visualisation of missing value')
plt.ylabel('missing value')
plt.xlabel('Rows')
plt.show()
````

####Drop rows with critical nulls (like Amount)

````python
df.dropna(subset=['Amount'], inplace=True)
````

####Fill non-critical nulls

````python
df['Age'].fillna(df['Age'].median(), inplace=True)
````

####Remove Duplicate Records

````python
df.drop_duplicates(inplace=True)
````

####Replace Values of Married_status

````python
df['Marital_Status'] = df['Marital_Status'].map({
    0: 'Unmarried',
    1: 'Married'
})
````

##Fix DataTypes

####Convert Amount & Orders to numeric

````python
df['Amount'] = pd.to_numeric(df['Amount'])
df['Orders'] = pd.to_numeric(df['Orders'])
````

####Convert categorical columns

````python
df['Gender'] = df['Gender'].astype('category')
df['Marital_Status'] = df['Marital_Status'].astype('category')
df['State'] = df['State'].astype('category')
df['Product_Category'] = df['Product_Category'].astype('category')
````

####Create Total_Sale Column

````python
df['Total_Sale'] = df['Amount'] * df['Orders']
````
